```python
from grape import Graph, GraphVisualizer
from grape.embedders import Node2VecGloVeEnsmallen
```

```python
# Load the graph from the CSV file
graph = Graph.from_csv(
    edge_path="../data/interim/edges.csv",
    sources_column="source",
    destinations_column="destination",
    directed=False,
)

graph
```

```python
# Generate Node2Vec embeddings using GloVeEnsmallen
print("Generating Node2Vec embeddings using Grape...")
embedding = Node2VecGloVeEnsmallen().fit_transform(graph)
```

```python
GraphVisualizer(graph).fit_and_plot_all(embedding)
```
